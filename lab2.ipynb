```
import numpy as np
import pandas as pd
import sklearn
from sklearn import preprocessing
from sklearn.model_selection import train_test_split
data = {
    'size': ['L', 'M',np.nan,'S','M','M','M','L','L','S'],
    'color': [np.nan, 'red', 'blue','black', 'black','yellow', 'pink', 'brown', 'black', 'white'],
    'gender': [np.nan,np.nan,'female','male',np.nan,'male',np.nan,'female','male','female'],
    'price': [np.nan, np.nan, 100.00, 120.00, 20.00,240.00, 140.88, 37.21, 99.99, 84.20],
    'weight': [488.00, 420.00, 120.00, 200.00, 220.00,180.00,240.00,np.nan,400.00, 390.00],
    'bought': ['no', 'no','no', 'yes','yes', 'no','yes', 'yes','yes', 'yes']
}
df_raw = pd.DataFrame(data=data)
df = df_raw.copy()
#df.info()
for col in ['size', 'color', 'gender', 'bought']:
    df[col] = df[col].astype('category')
df.info()
df.describe(include=['category']).T

df
df.dropna(thresh = (len(df.keys())-2), inplace=True)
df['price'].fillna(df['price'].mean(), inplace=True)
df['weight'].fillna(df['weight'].interpolate(), inplace=True)
df['size'].fillna('M', inplace=True)
df['color'].fillna('black', inplace=True)
df['gender'].fillna('male', inplace=True)
df['cost'] = pd.cut(df['price'], bins = 3, labels=('cheap', 'normal_price', 'expensive'))
df['bought_yes'] = df['bought'].apply(lambda x: 1 if 'yes' in x else 0)
df
df1=df[['price','cost']]
df2=df['bought_yes']

X_train, X_test, y_train, y_test = train_test_split(df1, df2, train_size=0.7)



#le = preprocessing.LabelEncoder()
#le.fit_transform(df['bought'])
#le.classes_
#pd.get_dummies(data=df)
#df = pd.get_dummies(data=df, drop_first=True)
#scaler = preprocessing.StandardScaler()

#df[['price', 'weight']] = scaler.fit_transform(df[['price', 'weight']])
df
```
