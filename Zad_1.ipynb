```
import numpy
import pandas as pd
import sklearn
from sklearn import preprocessing

data = {
    'size': ['L', 'M','S','S','M','M','M','L','L','S'],
    'color': ['green', 'red', 'blue','black', 'black','yellow', 'pink', 'brown', 'black', 'white'],
    'gender': ['female', 'male','female','male','female','male','female','female','male','female'],
    'price': [147.99, 54.50, 100.00, 120.00, 20.00,240.00, 140.88, 37.21, 99.99, 84.20],
    'weight': [488.00, 420.00, 120.00, 200.00, 220.00,180.00,240.00,300.00,400.00, 390.00],
    'bought': ['no', 'no','no', 'yes','no', 'no','yes', 'yes','yes', 'yes']
}
df_raw = pd.DataFrame(data=data)
df = df_raw.copy()
#df.info()
for col in ['size', 'color', 'gender', 'bought']:
    df[col] = df[col].astype('category')
df.info()
df.describe(include=['category']).T

le = preprocessing.LabelEncoder()
le.fit_transform(df['bought'])
le.classes_
pd.get_dummies(data=df)
df = pd.get_dummies(data=df, drop_first=True)
scaler = preprocessing.StandardScaler()
df[['price', 'weight']] = scaler.fit_transform(df[['price', 'weight']])
df
```
